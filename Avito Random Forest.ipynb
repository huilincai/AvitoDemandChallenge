```python
#import packages
import numpy as np 
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
np.random.seed(1234)
from sklearn import preprocessing
from sklearn import utils
```

```python
#import training data
training = pd.read_csv('newtrain.csv', encoding = 'utf-8') 
```

```python
training.head()
```

```python
#drop columns that will not be used (text columns) 
training=training.drop(['title', 'description', 'text_feat'], axis=1)
training.head()
```

```python
#split "training" dataset into train and test sets
training['is_train'] = np.random.uniform(0, 1, len(training)) <= .75
train, test = training[training['is_train']==True], training[training['is_train']==False]
```

```python
#train random forest classifier
clf = RandomForestClassifier(n_jobs=2, random_state=0)
features = (['Weekday', 'Weekd of Year', 'Day of Month', 'description_num_chars',
       'description_num_words', 'description_num_unique_words',
       'description_words_vs_unique', 'text_feat_num_chars',
       'text_feat_num_words', 'text_feat_num_unique_words',
       'text_feat_words_vs_unique', 'title_num_chars', 'title_num_words',
       'title_num_unique_words', 'title_words_vs_unique', 'user_id', 'region',
       'city', 'parent_category_name', 'category_name', 'price',
       'item_seq_number', 'user_type', 'image_top_1'])

y = train['deal_probability']
y=y.astype('int')

clf.fit(train[features], y)
```

```python
features
```

```python
#test classifier on test data (subset of training data)
pred = clf.predict(test[features])
```

```python
#calculate RMSE
#define RMSE
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
#apply RMSE function
rmse(pred, test['deal_probability']) 
```

```python
list(zip(train[features], clf.feature_importances_))
```

```python
#apply predictions to testing data for kaggle contest

#load and preparetesting data
testing = pd.read_csv('newtest.csv', encoding = 'utf-8') 
testing = testing.drop(['title', 'description', 'text_feat'], axis=1)

#create predictions for testing data
testing['rf_test_pred'] = clf.predict(testing[features])
testing['deal_probability'] = testing['rf_test_pred']
```
